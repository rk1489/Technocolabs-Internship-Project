Here, the 'Alcohol' feature is highly correlated with the 'Schooling' feature, So 'Schooling can be used to impute 'Alcohol.

Also, 'Hepatatis B' highly correlates with 'Diptheria' have null values in the same zone.

Also, other features can be imputed using features having good correlations with and null values not in the same zone.

```python
# Observing sctterplot between 'Alcohol' and 'Schooling'
sns.scatterplot(y=df['Alcohol'],x=df['Schooling'], hue=df['Developed'])
plt.show()
```

```python
# Imputing features with large no. of Null Values, here 'Alcohol'
def impute_to_Alcohol(cols):
    al=cols[0]
    sc=cols[1]
    if pd.isnull(al):
        if sc<=2.5:
            return 4.0
        elif 2.5<sc<=5.0:
            return 1.5
        elif 5.0<sc<=7.5:
            return 2.5
        elif 7.5<sc<=10.0:
            return 3.0
        elif 10.0<sc<=15:
            return 4.0
        elif sc>15:
            return 10.0
    else:
        return al
    
df['Alcohol'] = df[['Alcohol','Schooling']].apply(impute_to_Alcohol,axis=1)
```

```python
df['Alcohol'].isnull().sum()
```

```python
# Imputing the remaining nulls in 'Alcohol' with mean
df['Alcohol']=  df['Alcohol'].fillna(value=df['Alcohol'].mean())

df['Alcohol'].isnull().sum()
```

```python
# Imputing 'Polio' using 'Life Expectency'
sns.scatterplot(x=df['Life expectancy '],y=df['Polio'], hue=df['Developed'])
plt.show()
```

```python
def impute_to_Polio(cols):
    p=cols[0]
    l=cols[1]
    if pd.isnull(p):
        if l<=45:
            return 80.0
        elif 45<l<=50:
            return 67.0
        elif 50<l<=60:
            return 87.44
        elif 60<l<=70:
            return 91
        elif 70<l<=80:
            return 94.3
        elif l>80:
            return 95
    else:
        return p
    
df['Polio'] = df[['Polio','Life expectancy ']].apply(impute_to_Polio,axis=1)
```

```python
df['Polio'].isnull().sum()
```

```python
# Imputing 'Diphtheria' using 'Polio'
sns.scatterplot(x=df['Polio'],y=df['Diphtheria '], hue=df['Developed'])
plt.show()
```

```python
def impute_to_Diptheria(cols):
    d=cols[0]
    p=cols[1]
    if pd.isnull(d):
        if p<=10:
            return 75.0
        elif 10<p<=40:
            return 37.0
        elif 40<p<=45:
            return 40.0
        elif 45<p<=50:
            return 50.0
        elif 50<p<=60:
            return 55.0
        elif 60<p<=80:
            return 65.0
        elif p>80:
            return 90.0
    else:
        return d
df['Diphtheria ']=df[['Diphtheria ','Polio']].apply(impute_to_Diptheria,axis=1)
```

```python
df['Diphtheria '].isnull().sum()
```

```python
# Imputing 'Hepatitis B' using 'Diptheria'
sns.scatterplot(y=df['Hepatitis B'],x=df['Diphtheria '], hue=df['Developed'])
plt.show()
```

```python
def impute_to_Hepatatis(cols):
    hep=cols[0]
    dip=cols[1]
    if pd.isnull(hep):
        if dip<=15:
            return 75.0
        elif 15<dip<=30:
            return 20.0
        elif 30<dip<=45:
            return 38.0
        elif 45<dip<=60:
            return 43.0
        elif 60<dip<=80:
            return 63.0
        elif dip>80:
            return 88.4
    else:
        return hep
    
df['Hepatitis B'] = df[['Hepatitis B','Diphtheria ']].apply(impute_to_Hepatatis,axis=1)
```

```python
df['Hepatitis B'].isnull().sum()
```

```python
# Imputing  'BMI' using 'Life expactancy'
sns.scatterplot(x=df['Life expectancy '],y=df[' BMI '], hue=df['Developed'])
plt.show()
```

```python
def impute_to_BMI(cols):
    b=cols[0]
    l=cols[1]
    if pd.isnull(b):
        if l<=50:
            return 25.0
        elif 50<l<=60:
            return 25.0
        elif 60<l<=70:
            return 32.0
        elif 70<l<=80:
            return 46.8
        elif 80<l<=100:
            return 60.0
    else:
        return b
    
df[' BMI '] = df[[' BMI ','Life expectancy ']].apply(impute_to_BMI,axis=1)
```

```python
df[' BMI '].isnull().sum()
```

```python
# Imputing 'Total expenditure' using 'Alcohol'
sns.scatterplot(y=df['Total expenditure'],x=df['Alcohol'], hue=df['Developed'])
plt.show()
```

```python
def impute_to_Total_exp(cols):
    t=cols[0]
    a=cols[1]
    if pd.isnull(t):
        if a<=2.5:
            return 5.08
        elif 2.5<a<=5.0:
            return 6.0
        elif 5.0<a<=10.0:
            return 6.71
        elif 10.0<a<=12.5:
            return 6.9
        elif a>12.5:
            return 6.68
    else:
        return t
    
df['Total expenditure'] = df[['Total expenditure','Alcohol']].apply(impute_to_Total_exp,axis=1)        
```

```python
df['Total expenditure'].isnull().sum()
```

```python
# Imputing 'GDP' using 'Percentage Expenditure'
sns.scatterplot(x=df['percentage expenditure'],y=df['GDP'], hue=df['Developed'])
plt.show()
```

```python
def impute_to_GDP(cols):
    g=cols[0]
    p=cols[1]
    if pd.isnull(g):
        if p<=1250:
            return 1100.0
        elif 1250<p<=2500:
            return 1800.0
        elif 2500<p<=3750:
            return 2900.0
        elif 3750<p<=7500:
            return 3500.0
        elif 7500<p<=8750:
            return 4500.0
        elif 8750<p<=10000:
            return 5000.0
        elif 10000<p<=11250:
            return 5700.0
        elif 11250<p<=12500:
            return 7000.0
        elif 12500<p<=15000:
            return 8000.0
        elif 15000<p<=17500:
            return 9000.0
        elif p>17500:
            return 8500.0
    else:
        return g
    
df['GDP'] = df[['GDP','percentage expenditure']].apply(impute_to_GDP,axis=1)
```

```python
df['GDP'].isnull().sum()
```

```python
# Imputing 'Population' using 'Infant death'
sns.scatterplot(x=df['infant deaths'],y=df['Population'], hue=df['Developed'])
plt.show()
```

```python
def impute_to_population(cols):
    p=cols[0]
    i=cols[1]
    if pd.isnull(p):
        if i<=100:
            return 0.19*((10)**9)
        elif 100<i<=250:
            return 0.18*((10)**9)
        elif 250<i<=350:
            return 0.02*((10)**9)
        elif 350<i<=900:
            return 0.1*((10)**9)
        elif 900<i<=1100:
            return 0.18*((10)**9)
        elif 1100<i<=1250:
            return 0.05*((10)**9)
        elif 1250<i<=1500:
            return 0.19*((10)**9)
        elif 1500<i<=1750:
            return 0.05*((10)**9)
        elif i>1750:
            return 0.1*((10)**9)
    else:
        return p
    
df['Population'] = df[['Population','infant deaths']].apply(impute_to_population,axis=1)
```

```python
df['Population'].isnull().sum()
```

```python
# Imputing 'Thinness 1-19 years' using 'BMI'
sns.scatterplot(x=df[' BMI '],y=df[' thinness  1-19 years'], hue=df['Developed'])
plt.show()
```

```python
def impute_to_Thinness_19(cols):
    t=cols[0]
    b=cols[1]
    if pd.isnull(t):
        if b<=10:
            return 5.0
        elif 10<b<=20:
            return 10.0
        elif 20<b<=30:
            return 8.0
        elif 30<b<=40:
            return 6.0
        elif 40<b<=50:
            return 3.0
        elif 50<b<=70:
            return 4.0
        elif b>70:
            return 1.0
    else:
        return t
    
df[' thinness  1-19 years'] = df[[' thinness  1-19 years',' BMI ']].apply(impute_to_Thinness_19,axis=1)
```

```python
df[' thinness  1-19 years'].isnull().sum()
```

```python
# Imputing 'Thinness 5-9 years' using 'BMI'
sns.scatterplot(x=df[' BMI '],y=df[' thinness 5-9 years'], hue=df['Developed'])
plt.show()
```

```python
def impute_to_Thinness_9(cols):
    t=cols[0]
    b=cols[1]
    if pd.isnull(t):
        if b<=10:
            return 5.0
        elif 10<b<=20:
            return 10.0
        elif 20<b<=30:
            return 8.0
        elif 30<b<=40:
            return 6.0
        elif 40<b<=50:
            return 3.0
        elif 50<b<=70:
            return 4.0
        elif b>70:
            return 1.0
    else:
        return t
    
df[' thinness 5-9 years'] = df[[' thinness 5-9 years',' BMI ']].apply(impute_to_Thinness_19,axis=1)
```

```python
df[' thinness 5-9 years'].isnull().sum()
```

```python
# Imputing 'Income Composition of resources' using 'Life expactancy'
sns.scatterplot(x=df['Life expectancy '],y=df['Income composition of resources'], hue=df['Developed'])
plt.show()
```

```python
def impute_Income_compensation(cols):
    i=cols[0]
    l=cols[1]
    if pd.isnull(i):
        if l<=40:
            return 0.4
        elif 40<l<=50:
            return 0.42
        elif 50<l<=60:
            return 0.402
        elif 60<l<=70:
            return 0.54
        elif 70<l<=80:
            return 0.71
        elif l>80:
            return 0.88
    else:
        return i
        
df['Income composition of resources'] = df[['Income composition of resources','Life expectancy ']].apply(impute_Income_compensation,axis=1)      
```

```python
df['Income composition of resources'].isnull().sum()
```

```python
# Imputing 'Schooling' using 'Life expactancy'
sns.scatterplot(x=df['Life expectancy '],y=df['Schooling'], hue=df['Developed'])
plt.show()
```

```python
def impute_to_schooling(cols):
    s=cols[0]
    l=cols[1]
    if pd.isnull(s):
        if l<= 40:
            return 8.0
        elif 40<l<=44:
            return 7.5
        elif 44<l<50:
            return 8.1
        elif 50<l<=60:
            return 8.2
        elif 60<l<=70:
            return 10.5
        elif 70<l<=80:
            return 13.4
        elif l>80:
            return 16.5
    else:
        return s
    
df['Schooling'] = df[['Schooling','Life expectancy ']].apply(impute_to_schooling,axis=1)
```

### Training the Model and Predicting values

```python
# Checking the R Square Value Accuracy
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,predict_linear_reg)
```

##### Examining some other prediction models
##### Using the Lasso Regression model

```python
# Using Lasso Regression
from sklearn.linear_model import Lasso
```

```python
# Function to check Lasso accuracy for different alphas
def lasso_accuracy(alpha_value):
    reg = Lasso(alpha=alpha_value)
    reg.fit(X_train,y_train)
    predict = reg.predict(X_test)
    RMSE = mean_squared_error(y_test,predict)**(0.5)
    R2_score = r2_score(y_test,predict)
    print("Alpha =",alpha_value," RMSE =",RMSE," R2_score =",R2_score)
```

```python
# List of alpha values for accuracy check
alpha_value_list = [1, 0.01, 0.0001, 0.000001, 0.00000001, 0.0000000001]
```

```python
# Checking Lasso accuracy for different alphas
for i in alpha_value_list:
    lasso_accuracy(i)
```

```python
# Here, best alpha = 0.000001
lasso_reg = Lasso(alpha=0.000001)
```

```python
lasso_reg.fit(X_train,y_train)
```

```python
# Precicting the values using Lasso Regression
predict_lasso_reg = lasso_reg.predict(X_test)
```

```python
print(predict_lasso_reg[:10])
print()
print(y_test[:10].values)
```

```python
# Checking RMSE Accuracy of the model
mean_squared_error(y_test,predict_lasso_reg)**(0.5)
```

```python
# Checking the R Square Value Accuracy
r2_score(y_test,predict_lasso_reg)
```

##### Using the Ridge Regression model

```python
# Using Ridge Regression
from sklearn.linear_model import Ridge
```

```python
# Function to check Ridge accuracy for different alphas
def ridge_accuracy(alpha_value):
    reg = Ridge(alpha=alpha_value)
    reg.fit(X_train,y_train)
    predict = reg.predict(X_test)
    RMSE = mean_squared_error(y_test,predict)**(0.5)
    R2_score = r2_score(y_test,predict)
    print("Alpha =",alpha_value," RMSE =",RMSE," R2_score =",R2_score)
```

```python
# List of alpha values for accuracy check
alpha_value_list = [1, 0.01, 0.0001, 0.000001, 0.00000001, 0.0000000001]
```

```python
# Checking Ridge accuracy for different alphas
for i in alpha_value_list:
    ridge_accuracy(i)
```

```python
# Here, taking alpha = 0.000001
ridge_reg = Ridge(alpha=0.000001)
```

```python
ridge_reg.fit(X_train,y_train)
```

```python
# Precicting the values using Ridge Regression
predict_ridge_reg = ridge_reg.predict(X_test)
```

```python
print(predict_ridge_reg[:10])
print()
print(y_test[:10].values)
```

```python
# Checking RMSE Accuracy of the model
mean_squared_error(y_test,predict_ridge_reg)**(0.5)
```

```python
# Checking the R Square Value Accuracy
r2_score(y_test,predict_ridge_reg)
```

##### Finalising the Model to be used for Predictions

```python
# Having a look at accuracy of all models used above
def model_accuracy(model,predict_model):
    RMSE = mean_squared_error(y_test,predict_model)**(0.5)
    R2_score = r2_score(y_test,predict_model)
    print(model,":"," RMSE =",RMSE," R2_score =",R2_score)
    
predict_model_list = [['Linear Regression',predict_linear_reg],['Lasso',predict_lasso_reg],['Ridge',predict_ridge_reg]]

for i in predict_model_list:
    model_accuracy(i[0],i[1])
```

Here, almost all models are nearly the same but 'Lasso' seems to be a little better model for our data, so it will be the final model to be used for predictions.
